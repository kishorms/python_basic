```python
# import required module
from cryptography.fernet import Fernet
# GENTATE CODE
# key generation
key = Fernet.generate_key()

# string the key in a file
with open('/content/drive/MyDrive/Colab Notebooks/EkeyGen.txt', 'wb') as filekey:
    filekey.write(key)

# opening the key
with open('/content/drive/MyDrive/Colab Notebooks/EkeyGen.txt', 'rb') as filekey:
	  key = filekey.read()

# using the generated key
fernet = Fernet(key)

# opening the original file to encrypt
with open('/content/drive/MyDrive/Colab Notebooks/All_India_Index_july2019_20Aug2020_dec20_1_2.csv', 'rb') as file:
    original = file.read()
	
# encrypting the file
encrypted = fernet.encrypt(original)

# opening the file in write mode and
# writing the encrypted data
with open('/content/drive/MyDrive/Colab Notebooks/ENAll_India_Index_july2019_20Aug2020_dec20_1_2.csv', 'wb') as encrypted_file:
	  encrypted_file.write(encrypted)
   

```

```python
# Import libraries
import os
import time

# Insert the directory path in here
path = '/content/drive/MyDrive/Colab Notebooks/'

# Extracting all the contents in the directory corresponding to path
l_files = os.listdir(path)

# Iterating over all the files
for file in l_files:
  # Instantiating the path of the file
	file_path = f'{path}\\{file}'

	# Checking whether the given file is a directory or not
	if os.path.isfile(file_path):
		try:
			# Printing the file pertaining to file_path
			# os.startfile(file_path, 'print')
			print(f'Printing {file}')

			# Sleeping the program for 5 seconds so as to account the
	# steady processing of the print operation.
			time.sleep(5)
		except:
			# Catching if any error occurs and alerting the user
			print(f'ALERT: {file} could not be printed! Please check\
			the associated softwares, or the file type.')
	else:
		print(f'ALERT: {file} is not a file, so can not be printed!')
		
print('Task finished!')
```
