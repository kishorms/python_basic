```python
def write_key():
    """
    Generates a key and save it into a file
    """
    key = Fernet.generate_key()
    with open("key.key", "wb") as key_file:
        key_file.write(key)
```

```python
# import required module
from cryptography.fernet import Fernet
# GENTATE CODE
# key generation
key = Fernet.generate_key()

# string the key in a file
with open('/content/drive/MyDrive/Colab Notebooks/EkeyGen.txt', 'wb') as filekey:
    filekey.write(key)

# opening the key
with open('/content/drive/MyDrive/Colab Notebooks/EkeyGen.txt', 'rb') as filekey:
	  key = filekey.read()

# using the generated key
fernet = Fernet(key)

# opening the original file to encrypt
with open('/content/drive/MyDrive/Colab Notebooks/All_India_Index_july2019_20Aug2020_dec20_1_2.csv', 'rb') as file:
    original = file.read()
	
# encrypting the file
encrypted = fernet.encrypt(original)

# opening the file in write mode and
# writing the encrypted data
with open('/content/drive/MyDrive/Colab Notebooks/ENAll_India_Index_july2019_20Aug2020_dec20_1_2.csv', 'wb') as encrypted_file:
	  encrypted_file.write(encrypted)
   

```
